Cargamos las librerías necesarias para realizar el análisis

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import seaborn as sns
import matplotlib.pyplot as plt

import sys
import os
sys.path.append(os.path.abspath(".."))
```

Histograma de la distribución de Total puntos de combate

```python
df = pd.read_csv("../DataSet/PokemonData.csv")
df_sin_vacio = df["Type2"] = df["Type2"].fillna("None")

plt.figure(figsize=(10, 5))
sns.histplot(data=df_sin_vacio, x="Total", bins=30, kde=True)
plt.title("Distribución de puntuación total de Pokémon")
plt.xlabel("Total base stats")
plt.ylabel("Frecuencia")
plt.show()
```

Gráfico de barras de Pokémon por tipo primario

```python
plt.figure(figsize=(12, 6))
df["Type1"].value_counts().plot(kind="bar")
plt.title("Cantidad de Pokémon por tipo primario")
plt.xlabel("Tipo")
plt.ylabel("Número de Pokémon")
plt.xticks(rotation=45)
plt.show()
```
